```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import konlpy
from konlpy.tag import *
from nltk import Text
from wordcloud import WordCloud
```

```python
df=pd.read_csv(r"C:\Users\User\Downloads\data.csv")
df.tail()
```

```python
df=df[df['competition_rate']<1]
```

```python
len(df)
```

```python
text = " ".join(df['relKeyword'].values)
len(text), text[:50]
```

```python
okt = Okt() 
```

```python
nouns = okt.nouns(text)
```

```python
pos = okt.pos(text)
```

```python
verbs = [word for word, tagset in pos if tagset == "Noun"]
verbs[30:35]
```

```python
article = Text(nouns, name = "article")
datas = article.vocab()
```

```python
wc = WordCloud(font_path=r"C:\Users\User\Downloads\D2Coding.ttf", background_color="white")
```

```python
plt.figure(figsize=(20,5))
plt.imshow(wc.generate_from_frequencies(datas))
plt.axis('off')
plt.show()
```
