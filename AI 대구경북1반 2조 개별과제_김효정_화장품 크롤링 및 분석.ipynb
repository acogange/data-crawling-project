```python
#여러 페이지 리뷰 부르기
def review(page,page_size):
    url=f"https://brand.naver.com/n/v1/reviews/paged-reviews?page={page}&pageSize={page_size}&merchantNo=510133398&originProductNo=2538027516&sortType=REVIEW_RANKING"
    response=requests.get(url)
    datas = response.json()
    datas['contents']
    df = pd.DataFrame(datas['contents'])
    df.columns
    col=['id', 'reviewContent', 'createDate', 'repurchase','reviewScore',  'writerMemberId']

    return df[col]
```
